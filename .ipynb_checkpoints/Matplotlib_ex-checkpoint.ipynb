## Creating Plots Using the Functional Method

```python
# There are two ways in creating matplotlib plots: (1) Functional Method and (2) Object Oriented Method...

# 1) FUNCTIONAL METHOD IN CREATING MATPLOTLIB PLOTS.....
# NB: You don't need to use "plt.show()" in jupyter notebook
# what "plt.show()" does is to print out the plot instead of just showing the plot....


# Ex 1:-

plt.plot(x,y,'r-')               # 'r-' means the color red
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Title")
# plt.show()
```

### Creating Multi-Plots on the Same Canvas

```python
# Ex 2:- CREATING SUBPLOTS USING THE FUNCTIONAL METHOD....

plt.subplot(1,2,1)          # Here, 1 -> No. of rows, 2 -> No. of columns and 1 (3rd args) -> Plot Number.
plt.plot(x,y,'g')

plt.subplot(1,2,2)
plt.plot(y,x,'r')
```

## Creating Plots Using the Object Oriented Method

```python
# 2) THE OBJECT ORIENTED METHOD OF CREATING MATPLOTLIB PLOTS....

# Ex 3:-

fig = plt.figure()         # Creates a figure object...
```

```python
# The 1st arg -> means left of the axis - 10% from the left
# 2nd args -> means Bottom of the axis - 10% from the bottom
# 3rd args -> means Width of the axis - 75% of the canvas size width
# 4th args -> means Height of the axis - 75% of the canvas size height

fig = plt.figure()  
axis = fig.add_axes([0.1,0.1,0.75,0.75])           # "add_axes" is a method...
```

```python
fig = plt.figure()  
axis = fig.add_axes([0.1,0.1,0.75,0.75])       # The list in the "add_axes()" method defines the location of the axis

axis.plot(x,y)
axis.set_xlabel("X Axis")
axis.set_ylabel("Y Axis")
axis.set_title("Title")
```

```python
# Ex 4:-

fig = plt.figure()  

axis1 = fig.add_axes([0.1,0.1,0.75,0.75])   
axis2 = fig.add_axes([0.2,0.45,0.4,0.35])
```

```python
# Ex 5:- Moving the axis2 to the bottom right corner of the plot(axis1)

fig = plt.figure()  
axis1 = fig.add_axes([0.1,0.1,0.75,0.75])
axis2 = fig.add_axes([0.43,0.15,0.4,0.35])
```

```python
# Ex 6:- Plotting (x,y) on both axes and calling the figure object to show it

fig = plt.figure()  
axis1 = fig.add_axes([0.1,0.1,0.75,0.75])
axis2 = fig.add_axes([0.2,0.45,0.35,0.30])

axis1.plot(x,y)
axis1.set_title("LARGER PLOT")

axis2.plot(y,x,"r")
axis2.set_title("SMALLER PLOT")
```

### Creating subplots() using Object Oriented Method
subplots() differs from ordinary plots because we can specify the number of rows & cols.... 

```python
# Ex 7i:-

fig,axes = plt.subplots(nrows=1,ncols=2)       # We did a tuple unpacking by joining both "fig" & "axes" together on one line...
```

```python
# Ex 7ii:-

fig,axes = plt.subplots(nrows=3,ncols=4)
```

```python
# Ex 7iii:- Fixing the joinings of the plots, we use the -> tight_layout() method

fig,axes = plt.subplots(nrows=3,ncols=4)
plt.tight_layout() 
```

```python
# Ex 8:-

fig,axes = plt.subplots(nrows=1,ncols=2)
for showing_axis_points in axes:
    showing_axis_points.plot(x,y)
```

```python
# Ex 9i:- Specifying the particular axis to include data points on the plot, we indicate the index

fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(y,x)                       # Shows data points...
```

```python
# Ex 9ii:- Including data points on both plots and indicating their titles....

fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(y,x)
axes[0].set_title("First Plot")

axes[1].plot(x,y)
axes[1].set_title("Second Plot")
```

## Figure Size, Aspect Ratio and DPI
Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

    -> figsize is a tuple of the width and height of the figure in inches
    -> dpi is the dots-per-inch (pixel per inch).

```python
# figsize=(3,2) -> means the width,height of the figure in inches..
# dpi=110 -> means resolution of the figure in "dots per inch" or "pixel per inch"...

# Ex 10i:-

fig = plt.figure(figsize=(3,2), dpi=110)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# Ex 10ii:-

fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# SUBPLOTS....

# Ex 11i:-
fig,axes = plt.subplots(figsize=(8,2))

axes.plot(x,y)
```

```python
# Ex 11ii:-

fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")

plt.tight_layout()
```

```python
# Ex 12v:- Adding a "legend"
# Specifying *args "loc=0" tells matplotlib to find the best location to put our legend... 
# You could also specify different location codes/location string. e.g -> loc(1)/"upper right"...

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label="X Squared")
ax.plot(x, x**3, label="X Cubed")

ax.legend(loc=0)
```

```python
# Ex 12vi:-

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label="X Squared")
ax.plot(x, x**3, label="X Cubed")

ax.legend(loc=9)
```

```python
# Ex 12vii:- USING A TUPLE() TO SPECIFY A LEGEND LOCATION..

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label="X Squared")
ax.plot(x, x**3, label="X Cubed")

ax.legend(loc=(0.2,0.23))      
```

## Plot Appearance - Setting colors, linewidths, linetypes & drawstyle
Matplotlib gives you a lot of options for customizing colors, linewidths, and linetypes.

There is the basic MATLAB like syntax which I would suggest you avoid using for more clarity sake.

### Colors with MatLab like syntax
With matplotlib, we can define the colors of lines and other graphical elements in a number of ways. First of all, we can use the MATLAB-like syntax where 'b' means blue, 'g' means green, etc. The MATLAB API for selecting line styles are also supported: where, for example, 'b.-' means a blue line with dots:

```python
# Ex 13:- MATLAB style line color and style 

fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

### Colors with the color= parameter
We can also define colors by their names or RGB hex codes and optionally provide an alpha value using the `color` and `alpha` keyword arguments. Alpha indicates opacity.

```python
# Ex 14i:-

fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")     
```

```python
# Ex 14ii:- Using # RGB Hexadecimal codes for colors

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="#444444")
```

```python
# Ex 14iii:- Using common color names

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="green")
```

```python
# Ex 15i:- Using linewidths -> "linewidth creates a Line 2D image"...


fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="purple", linewidth=2)
```

```python
# Ex 15ii:- Using the "alpha" argument -> helps specify how transparent the line is...
# Note that the shorthand for linewidth is --> "lw"


fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="purple", lw=3, alpha=0.6)
```

```python
# Ex 15iii:- Using the "linestyle" argument -> helps define the style of the line. Such as "--" or "..." or ":"
# Note that the shorthand for linestyle is --> "ls"

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="purple", lw=3, alpha=0.6, linestyle="--")
```

```python
# Ex 15iv:- Using the "drawstyle" argument -> helps define the style of the line. Such as "--" or "..." or ":"
# Note that the shorthand for drawstyle is --> "ds"

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="purple", lw=3, alpha=0.6, ds="steps")
```

```python
# Ex 16ii:- Using the "markersize" -> Makes the marker bigger/smaller...

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="purple", lw=3, alpha=0.6, marker="*", markersize=7)
```

### Line and marker styles

### Control over axis appearance - Plot range
We can configure the ranges of the axes using the set_ylim and set_xlim methods in the axis object, or axis('tight') for automatically getting "tightly fitted" axes ranges.

```python
# Ex 18i:- Show plot between 0 and 1 on the x-axis, & 1 and 2 on y-axis....
# ax.set_ylim([0, 2]) -> means 0 is: lower bound, 2: upper bound. It also applies to x-axis based on the parameters given..
# Lower bound on y-axis -> means the lowest data point on y is:"0.00" and highest data point on the same y is:"2.00" 

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="purple", lw=2.5, alpha=0.6, ls="dashed")
ax.set_ylim([0, 2])
ax.set_xlim([0, 1])
```

```python
# Ex 18ii:- 

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```
